```python
#Level by quaters
df['quater with year'] = pd.PeriodIndex(df.Date, freq='Q')
df['quater']=df['quater with year'].dt.quarter
df.head()
```

```python
df_quater1=df[df.quater==1]
# df_quater1=df_quater1["Weekly_Sales" ]
df_quater2=df[df.quater==2]
# df_quater2=df_quater2["Weekly_Sales" ]
df_quater3=df[df.quater==3]
# df_quater3=df_quater3["Weekly_Sales" ]
df_quater4=df[df.quater==4]
# df_quater4=df_quater4["Weekly_Sales" ]
# #df_quater1.transpose()
```

```python
# OneWayAnovaTestStats, OneWayAnovaPVal=f_oneway(df_quater1,
#                                                df_quater2,
#                                                df_quater3,
#                                                df_quater4)

# print("P value is", OneWayAnovaPVal)
# Siglevel=0.05
# if OneWayAnovaPVal<=Siglevel:
#     print("We reject H0. There exists a difference in the weekly sales level for different quater.")
# else:
#     print("We do not reject H0. We cannot conclude that there exists a difference in the weekly sales level for different quater.")
```

```python
store=["Quater"]
quater1=["quater 1"]
quater2=["quater 2"]
quater3=["quater 3"]
quater4=["quater 4"]
for i in range(1,46):
  rand=i
  store.append("store {}".format(rand))
  df_quater1_loc=df_quater1[df_quater1.Store==rand]
  df_quater1_loc_ws=df_quater1_loc.Weekly_Sales
  avg_df_quater1_loc_ws=np.mean(df_quater1_loc_ws)
  quater1.append(avg_df_quater1_loc_ws)

  df_quater2_loc=df_quater2[df_quater2.Store==rand]
  df_quater2_loc_ws=df_quater2_loc.Weekly_Sales
  avg_df_quater2_loc_ws=np.mean(df_quater2_loc_ws)
  quater2.append(avg_df_quater2_loc_ws)


  df_quater3_loc=df_quater3[df_quater3.Store==rand]
  df_quater3_loc_ws=df_quater3_loc.Weekly_Sales
  avg_df_quater3_loc_ws=np.mean(df_quater3_loc_ws)
  quater3.append(avg_df_quater3_loc_ws)

  df_quater4_loc=df_quater4[df_quater4.Store==rand]
  df_quater4_loc_ws=df_quater4_loc.Weekly_Sales
  avg_df_quater4_loc_ws=np.mean(df_quater4_loc_ws)
  quater4.append(avg_df_quater4_loc_ws)

print(quater1)
print(quater2)
print(quater3)
print(quater4)


oneway_list=[quater1,quater2,quater3,quater4]
oneway_df=pd.DataFrame(oneway_list,columns=store)
oneway_df
```

```python
OneWayAnovaTestStats, OneWayAnovaPVal=f_oneway(quater1[1:],
                                               quater2[1:],
                                               quater3[1:],
                                               quater4[1:])

print("P value is", OneWayAnovaPVal)
Siglevel=0.05
if OneWayAnovaPVal<=Siglevel:
    print("We reject H0. There exists a difference in the total weekly sales level for different quater.")
else:
    print("We do not reject H0. We cannot conclude that there exists a difference in the total weekly sales level for different quater.")
```
